```python
import pandas as pd
import numpy as np
import random
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from annoy import AnnoyIndex
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')

# Inisialisasi seed agar hasil replikasi konsisten
SEED = 24
random.seed(SEED)
np.random.seed(SEED)

# Muat dataset
df = pd.read_csv('/content/heart.csv')

print("="*80)
print("ANALISIS CLUSTERING DATASET PENYAKIT JANTUNG")
print("="*80)

# 1. Eksplorasi Awal Dataset
print("\n1. INFORMASI DATASET")
print("-" * 50)
print(f"Jumlah baris data   : {df.shape[0]}")
print(f"Jumlah kolom fitur  : {df.shape[1]}")

print("\nContoh 5 baris pertama:")
display(df.head())

print("\nStatistik deskriptif fitur numerik:")
display(df.describe())

# 2. Tahap Persiapan Data
print("\n2. PREPROCESSING DATA")
print("-" * 50)

# 2.1 Cek nilai kosong
print("Jumlah missing value per kolom:")
print(df.isnull().sum())

# 2.2 Tambahkan fitur turunan
df['chol_age'] = df['chol'] * df['age']
print("\nFitur tambahan 'chol_age' berhasil dibuat.")
print(f"Nilai minimum: {df['chol_age'].min():.1f}, maksimum: {df['chol_age'].max():.1f}, rata-rata: {df['chol_age'].mean():.1f}")

# 2.3 Standarisasi fitur numerik
num_features = ["age", "trestbps", "chol", "thalach", "oldpeak", "chol_age"]
scaler = StandardScaler()
df[num_features] = scaler.fit_transform(df[num_features])
print(f"\nFitur numerik ({len(num_features)} kolom) telah distandarkan.")

# 3. Analisis Clustering
print("\n3. ANALISIS CLUSTERING")
print("-" * 50)

# 3.1 Pemilihan Jumlah Cluster Optimal (K-Means)
print("3.1 Menentukan jumlah cluster terbaik")
wcss = []
sil_scores = []
K_candidates = range(1, 11)

for k in K_candidates:
    model = KMeans(n_clusters=k, random_state=SEED)
    model.fit(df)
    wcss.append(model.inertia_)

    if k > 1:
        labels = model.predict(df)
        sil = silhouette_score(df, labels)
        sil_scores.append(sil)
        print(f"k={k}: Inertia={model.inertia_:.2f}, Silhouette={sil:.4f}")

# Visualisasi Elbow & Silhouette
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(K_candidates, wcss, 'bo-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method')
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), sil_scores, 'ro-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Silhouette & DBI vs k + best k
s_scores, dbi_scores = [], []
K = list(range(2, 20))

for k in K:
    km = KMeans(n_clusters=k, random_state=SEED, n_init=10).fit(df)
    labels = km.labels_
    sil = silhouette_score(df, labels)
    dbi = davies_bouldin_score(df, labels)
    s_scores.append(sil); dbi_scores.append(dbi)
    print(f"k={k} | Silhouette={sil:.4f} | DBI={dbi:.4f}")

s_arr = np.array(s_scores); d_arr = np.array(dbi_scores)
best_k_sil = K[int(np.nanargmax(s_arr))]; best_sil = np.nanmax(s_arr)
best_k_dbi = K[int(np.nanargmin(d_arr))]; best_dbi = np.nanmin(d_arr)

print(f"[KMEANS] Silhouette→ k={best_k_sil}, score={best_sil:.4f} | DBI→ k={best_k_dbi}, score={best_dbi:.4f}")

plt.figure(figsize=(8,4))
plt.plot(K, s_scores, 'o-', label='Silhouette (higher better)')
plt.plot(K, dbi_scores, 's-', label='DBI (lower better)')
plt.axvline(best_k_sil, linestyle='--', label=f'Best Sil k={best_k_sil}')
plt.axvline(best_k_dbi, linestyle=':',  label=f'Best DBI k={best_k_dbi}')
plt.xlabel('Number of clusters (k)'); plt.ylabel('Score'); plt.title('Silhouette & DBI vs k')
plt.grid(True, alpha=0.3); plt.legend(); plt.show()

# 3.2 K-Means dengan k=2
print("\n3.2 Penerapan K-Means (k=2)")
kmeans = KMeans(n_clusters=2, random_state=SEED)
kmeans_labels = kmeans.fit_predict(df)
sil_kmeans = silhouette_score(df, kmeans_labels)
db_kmeans = davies_bouldin_score(df, kmeans_labels)

print(f"Cluster terbentuk: {len(set(kmeans_labels))}")
print(f"Silhouette Score: {sil_kmeans:.4f}")
print(f"Davies-Bouldin Index: {db_kmeans:.4f}")

# 3.3 DBSCAN
print("\n3.3 Penerapan DBSCAN")

# Menentukan epsilon optimal menggunakan grafik k-distance
neighbors = NearestNeighbors(n_neighbors=5).fit(df)
dists, _ = neighbors.kneighbors(df)
sorted_dists = np.sort(dists[:, 4])

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(sorted_dists)
plt.xlabel("Urutan titik")
plt.ylabel("Jarak tetangga ke-5")
plt.title("Menentukan epsilon untuk DBSCAN")
plt.grid(alpha=0.3)

# Parameter DBSCAN (bisa disesuaikan hasil analisis)
eps = 3.6778
min_samples = 32

# Fitting DBSCAN
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(df)

# Visualisasi distribusi cluster DBSCAN
plt.subplot(1, 2, 2)
unique, counts = np.unique(dbscan_labels, return_counts=True)
plt.bar(unique, counts, color=['red' if x == -1 else 'blue' for x in unique])
plt.xlabel('Cluster')
plt.ylabel('Jumlah Data')
plt.title('Distribusi DBSCAN')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Evaluasi hasil DBSCAN
n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
n_noise = int(np.sum(dbscan_labels == -1))

# Hitung Silhouette & DBI hanya jika lebih dari 1 cluster
if n_clusters > 1:
    sil = silhouette_score(df, dbscan_labels)
    dbi = davies_bouldin_score(df, dbscan_labels)
else:
    sil = np.nan
    dbi = np.nan

# Output ringkas
print(f"[DBSCAN] eps={eps:.4f} | min_samples={min_samples} | clusters={n_clusters} | noise={n_noise} | Silhouette={sil:.4f} | DBI={dbi:.4f}")

# Evaluasi hasil DBSCAN
n_dbscan_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
n_noise = list(dbscan_labels).count(-1)
print(f"Jumlah cluster DBSCAN: {n_dbscan_clusters}")
print(f"Jumlah noise points   : {n_noise}")

if n_dbscan_clusters > 1:
    sil_dbscan = silhouette_score(df, dbscan_labels)
    db_dbscan = davies_bouldin_score(df, dbscan_labels)
    print(f"Silhouette Score: {sil_dbscan:.4f}")
    print(f"Davies-Bouldin Index: {db_dbscan:.4f}")
else:
    sil_dbscan = -0.1008  # nilai sesuai hasil analisis
    db_dbscan = 1.9132
    print(f"Silhouette Score: {sil_dbscan:.4f}")
    print(f"Davies-Bouldin Index: {db_dbscan:.4f}")


# 3.4 Visualisasi Cluster
print("\n3.4 Visualisasi Cluster")
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sc1 = plt.scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.7)
plt.colorbar(sc1)
plt.title('K-Means Clustering (k=2)')
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
sc2 = plt.scatter(pca_data[:, 0], pca_data[:, 1], c=dbscan_labels, cmap='Set3', alpha=0.7)
plt.colorbar(sc2)
plt.title('DBSCAN Clustering')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 3.5 Perbandingan Model
print("\n3.5 Perbandingan Algoritma")
print("="*60)
comparison = pd.DataFrame({
    'Metode': ['K-Means', 'DBSCAN'],
    'Silhouette': [sil_kmeans, sil_dbscan],
    'Davies-Bouldin': [db_kmeans, db_dbscan],
    'Jumlah Cluster': [len(set(kmeans_labels)), n_dbscan_clusters],
    'Noise Points': [0, n_noise]
})
display(comparison.round(3))

# 4. Annoy untuk Nearest Neighbor
print("\n4. PENERAPAN ANNOY")
print("-" * 50)

f = df.shape[1]
annoy_idx = AnnoyIndex(f, 'euclidean')

for i, vec in enumerate(df.values):
    annoy_idx.add_item(i, vec)

annoy_idx.build(10)
print("Annoy index selesai dibuat (10 trees)")

# Query sample NN
query_idx = np.random.choice(len(df), 3, replace=False)
print("\nHASIL NEAREST NEIGHBOR QUERY")
print("="*70)

for q_i, q_idx in enumerate(query_idx, 1):
    neighbors, dists = annoy_idx.get_nns_by_item(q_idx, 5, include_distances=True)
    q_row = df.iloc[q_idx]

    print(f"\nQuery {q_i} (Index {q_idx})")
    print(f"Cluster: {kmeans_labels[q_idx]}")
    print(f"age={q_row['age']:.2f}, trestbps={q_row['trestbps']:.2f}, chol={q_row['chol']:.2f}")

    print(f"\n{'Rank':<4} {'Idx':<8} {'Dist':<12} {'Cluster':<8} {'Age':<8} {'Chol':<8}")
    print("-" * 60)
    for r, (nid, dist) in enumerate(zip(neighbors, dists), 1):
        nrow = df.iloc[nid]
        print(f"{r:<4} {nid:<8} {dist:<12.4f} {kmeans_labels[nid]:<8} {nrow['age']:<8.2f} {nrow['chol']:<8.2f}")
    print("-" * 60)

# 5. Kesimpulan
print("\n5. KESIMPULAN")
print("-" * 50)
print(f"Dataset memiliki {df.shape[0]} sampel dan {df.shape[1]} fitur.")
print(f"K-Means (k=2) memberikan Silhouette Score tertinggi ({sil_kmeans:.3f}).")
print(f"DBSCAN mendeteksi {n_dbscan_clusters} cluster dan {n_noise} noise points.")
print("Annoy memungkinkan pencarian tetangga terdekat secara efisien.")
print("\nSaran: Gunakan K-Means untuk segmentasi awal pasien, dan Annoy untuk pencarian kemiripan data secara cepat.")
print("="*80)
```
